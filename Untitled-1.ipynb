```python
df=df.drop('Other',axis=1)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df = df.sort_values('Date')
```

```python
df.set_index('Date', inplace=True)
```

Handling missing datas and outliers

```python
full_range = pd.date_range(df.index.min(), df.index.max(), freq='D')
df = df.reindex(full_range)
df = df.interpolate(method='time')
```

```python
cols = ['Local Route','Light Rail','Peak Service','Rapid Route','School']
Q1 = df[cols].quantile(0.25)
Q3 = df[cols].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = ((df[cols] < lower_bound) | (df[cols] > upper_bound)).sum()
print("Outliers detected per column:\n", outliers)


plt.figure(figsize=(10,5))
plt.boxplot(df[cols])
plt.xticks(range(1,6), cols, rotation=30)
plt.title("Outlier Detection (Before Handling)")
plt.show()


for c in cols:
    median_val = df[c].median()
    df.loc[(df[c] < lower_bound[c]) | (df[c] > upper_bound[c]), c] = median_val

#check
outliers_after = ((df[cols] < lower_bound) | (df[cols] > upper_bound)).sum()
print("\nOutliers per column after handling:\n", outliers_after)


plt.figure(figsize=(10,5))
plt.boxplot(df[cols])
plt.xticks(range(1,6), cols, rotation=30)
plt.title("Outlier Detection (After Handling)")
plt.show()
```
